```python
import re
import collections
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans, MiniBatchKMeans, DBSCAN

from gensim.models import doc2vec
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
newsgroups_train = fetch_20newsgroups(subset='train')
```

```python
np.shape(newsgroups_train.data)
```

```python
# Convert text to lower-case and strip punctuation/symbols from words
def normalize_text(text):
    norm_text = text.lower()

    # Replace breaks with spaces
    norm_text = re.sub(r'\n', ' ', norm_text)
    norm_text = re.sub(r'[^a-z0-9, \.]', '', norm_text)
    norm_text = re.sub(r'\b[0-9,\.]{1,}\b', '<NUMBER>', norm_text)

    # Pad punctuation with spaces on both sides
    for char in ['.', '"', ',', '(', ')', '!', '?', ';', ':']:
        norm_text = norm_text.replace(char, ' ' + char + ' ')

    return norm_text
```

```python
docs = [doc2vec.TaggedDocument(normalize_text(text).split(' '), [i]) for i, text in enumerate(newsgroups_train.data)]
model = doc2vec.Doc2Vec(docs, size=100, window=8, min_count=5, workers=4, iter=20, seed=0)
```

```python
# get tfidf of the topics
def topic_keywords(topic, text):
    df = pd.DataFrame({'topic':pd.Series(topic), 'text':pd.Series([normalize_text(t) for t in text])}, columns=['topic', 'text'])
    topic_text = df.groupby('topic').agg({'text': lambda x: ' '.join(x) })
    vec = TfidfVectorizer(max_features=5000, stop_words='english', strip_accents='unicode', max_df=0.5, min_df=0.05)
    X = vec.fit_transform(topic_text.text)
    features = vec.get_feature_names()
    return df, [sorted(zip(features, row), key=lambda a: a[1], reverse=True)[0:100] for row in X.toarray()]
```

```python
kmean = MiniBatchKMeans(20, max_iter=200)
Y = kmean.fit_predict(model.docvecs)
dist = kmean.fit_transform(model.docvecs)
```

```python
df, keywords = topic_keywords(Y, newsgroups_train.data)
```

```python
for text in df.sort_values('topic').text[0:5]:
    print(text)
    print('####\n')
```
